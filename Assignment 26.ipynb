```python
Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Ans. Matplotlib is a popular data visualization library in Python that allows users to create a wide range of static, animated, and 
interactive visualizations. It was originally created by John D. Hunter in 2003 as an open-source project, and it has
since become one of the most widely used data visualization tools in the scientific Python ecosystem.

Matplotlib is used for various tasks such as exploratory data analysis, creating publication-quality figures, and generating 
charts and graphs for reports and presentations. It is also commonly used in the scientific community for 
visualizing data from experiments and simulations.

Here are five plots that can be plotted using the Pyplot module of Matplotlib:

Line plot: A line plot is a basic plot type that displays data as a series of points connected by straight lines. It is useful 
for showing trends over time or between variables.

Scatter plot: A scatter plot is a plot type that displays data as a collection of points. It is useful for identifying patterns 
and relationships between variables.

Bar plot: A bar plot is a plot type that displays data as a series of rectangular bars. It is useful for comparing values between 
different categories.

Histogram: A histogram is a plot type that displays the distribution of a dataset. It is useful for understanding the frequency
distribution of continuous variables.

Box plot: A box plot is a plot type that displays the distribution of a dataset using quartiles. It is useful for identifying
outliers and understanding the spread of a dataset.
```

```python
Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

Ans. A scatter plot is a type of plot used to display the relationship between two continuous variables. It involves
plotting one variable on the x-axis and another variable on the y-axis, and then displaying the data as a collection of points.

Here's the code to generate data for x and y and plot a scatter plot using Matplotlib's Pyplot module:
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)
plt.title('Scatter plot of X and Y')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

Ans. The subplot() function in Matplotlib is used to create multiple plots in a single figure. It allows users to
display several plots side by side or stacked on top of each other. The subplot() function takes three arguments: the number of rows, the
number of columns, and the index of the plot to be created.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the data for four lines
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure with 2 rows and 2 columns of subplots
fig, axs = plt.subplots(2, 2)

# Plot the first line on the first subplot
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')

# Plot the second line on the second subplot
axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

# Plot the third line on the third subplot
axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

# Plot the fourth line on the fourth subplot
axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

# Add a suptitle for the entire figure
fig.suptitle('Four Line Plots')

# Display the figure
plt.show()
```

```python
Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

Ans. A bar plot, also known as a bar chart, is a type of graph that represents categorical data with rectangular bars.
The height or length of each bar is proportional to the value it represents. Bar plots are used to compare and display
data between different categories.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the data for the bar plot
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a figure with a single subplot for the vertical bar plot
fig, ax = plt.subplots()
ax.bar(company, profit)
ax.set_xlabel('Company')
ax.set_ylabel('Profit')
ax.set_title('Vertical Bar Plot')

# Create a figure with a single subplot for the horizontal bar plot
fig, ax = plt.subplots()
ax.barh(company, profit)
ax.set_xlabel('Profit')
ax.set_ylabel('Company')
ax.set_title('Horizontal Bar Plot')

# Display the figures
plt.show()
```

```python
Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

Ans. A box plot, also known as a box-and-whisker plot, is a type of graph used to display the distribution of a dataset.
It summarizes the data using five statistics: the minimum value, the first quartile, the median, the third quartile, and the maximum 
value. It is useful for identifying outliers and comparing the distribution of multiple datasets.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the data for the box plot
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create a figure with a single subplot for the box plot
fig, ax = plt.subplots()
ax.boxplot([box1, box2])
ax.set_xticklabels(['Box 1', 'Box 2'])
ax.set_ylabel('Values')
ax.set_title('Box Plot')

# Display the figure
plt.show()
```

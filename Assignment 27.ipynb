```python
Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.
Ans. Seaborn is a Python data visualization library that is built on top of Matplotlib. It provides a higher-level
interface for creating informative and attractive statistical graphics. Here are five plots that can be 
plotted using Seaborn and their uses:

Scatter plot: A scatter plot is a plot of two variables showing their correlation. Seaborn provides a high-level interface
for creating scatter plots with different markers, colors, and sizes. Scatter plots are useful for visualizing patterns in
data, identifying trends, and detecting outliers.

Histogram: A histogram is a plot that shows the distribution of a continuous variable. Seaborn provides a high-level interface
for creating histograms with different bin sizes, colors, and styles. Histograms are useful for exploring the shape of the data 
distribution, identifying outliers, and detecting skewness.

Line plot: A line plot is a plot of one or more variables showing their trends over time or other ordered categories. Seaborn
provides a high-level interface for creating line plots with different colors, markers, and styles. Line plots are useful for
visualizing time series data, trends, and changes over time.

Bar plot: A bar plot is a plot that shows the relationship between a categorical variable and a continuous variable. Seaborn 
provides a high-level interface for creating bar plots with different colors, order, and orientation. Bar plots are useful
for comparing the relative sizes of groups, showing patterns over time or space, and visualizing proportions.

Heatmap: A heatmap is a plot that shows the relationship between two categorical variables or a categorical and a continuous 
variable using colors. Seaborn provides a high-level interface for creating heatmaps with different color schemes, annotations,
and scales. Heatmaps are useful for visualizing patterns, correlations, and relationships in data, especially in large datasets.

Overall, Seaborn provides a wide range of plots for visualizing different types of data and exploring their relationships.
```

```python
Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.
```

```python
import seaborn as sns

# Load the "fmri" dataset
fmri = sns.load_dataset("fmri")

# Create a line plot with different events and regions
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri)

# Set plot title and axis labels
plt.title("Signal over Time by Event and Region")
plt.xlabel("Timepoint")
plt.ylabel("Signal")

# Show the plot
plt.show()
```

```python
Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.
```

```python
import seaborn as sns

# Load the "titanic" dataset
titanic = sns.load_dataset("titanic")

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 6))

# Create a box plot for "age" vs. "pclass"
sns.boxplot(x="pclass", y="age", data=titanic, ax=ax1)
ax1.set_title("Age by Passenger Class")

# Create a box plot for "fare" vs. "pclass"
sns.boxplot(x="pclass", y="fare", data=titanic, ax=ax2)
ax2.set_title("Fare by Passenger Class")

# Show the plot
plt.show()
```

```python
Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Load the diamonds dataset
diamonds = sns.load_dataset('diamonds')

# Create a histogram with hue = 'cut'
sns.histplot(data=diamonds, x='price', hue='cut', multiple='stack')

# Add labels and title
plt.xlabel('Price')
plt.ylabel('Count')
plt.title('Histogram of Diamond Prices by Cut')

# Show the plot
plt.show()
```

```python
Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.
```

```python
Que 6: Use the "flights" dataset from seaborn to plot a heatmap.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Load the flights dataset
flights_data = sns.load_dataset('flights')

# Reshape the data into a pivot table
flights_pivot = flights_data.pivot('month', 'year', 'passengers')

# Create a heatmap using Seaborn
sns.heatmap(flights_pivot, annot=True, fmt='d')

# Add title and axis labels
plt.title('Passenger Count by Year and Month')
plt.xlabel('Year')
plt.ylabel('Month')

# Show the plot
plt.show()
```

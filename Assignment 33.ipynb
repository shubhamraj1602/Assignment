```python
Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with
an example.

Ans. The Probability Mass Function (PMF) and Probability Density Function (PDF) are both ways of describing the probability
distribution of a random variable.

The PMF is a function that gives the probability that a discrete random variable takes on a specific value. For example,
consider a fair six-sided die. The PMF of the outcome of rolling the die can be expressed as:

Die Roll	Probability
1	1/6
2	1/6
3	1/6
4	1/6
5	1/6
6	1/6Here, we have a discrete random variable, the outcome of rolling a die, and the PMF gives the probability that
the random variable takes on a specific value (the die roll).

On the other hand, the PDF is a function that describes the probability density of a continuous random variable over a 
range of values. For example, consider the normal distribution with mean 0 and standard deviation 1. The PDF of this
distribution is given by the formula:

f(x) = (1 / (sqrt(2 * pi) * sigma)) * exp(-(x - mu)^2 / (2 * sigma^2))

where mu is the mean and sigma is the standard deviation. This function gives the probability density of the normal 
distribution at any given point x.

Here, we have a continuous random variable, and the PDF gives the probability density (probability per unit of measurement) 
at each point within a range of values. For example, the probability that the random variable takes on a value between -1 and 
1 is given by integrating the PDF between -1 and 1.

In summary, the PMF is used for discrete random variables and gives the probability of specific values, while 
the PDF is used for continuous random variables and gives the probability density at each point within a range of values.
```

```python
Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?
Ans. The Cumulative Distribution Function (CDF) is a function that gives the cumulative probability of a random 
variable taking on a value less than or equal to a given value. In other words, it is the probability that the 
random variable is less than or equal to a particular value.

For a discrete random variable, the CDF can be expressed as a step function that increases at each point where the
PMF has a nonzero probability. For a continuous random variable, the CDF is a continuous function.

For example, consider a fair six-sided die. The CDF of the outcome of rolling the die can be expressed as:

Die Roll	Probability	Cumulative Probability
1	1/6	1/6
2	1/6	1/3
3	1/6	1/2
4	1/6	2/3
5	1/6	5/6
6	1/6	1
Here, we have a discrete random variable, the outcome of rolling a die, and the CDF gives the cumulative probability 
that the random variable takes on a value less than or equal to a specific value (the die roll).

The CDF is used in probability and statistics to analyze and understand the distribution of random variables. It can 
be used to find the probability that a random variable takes on a value within a certain range, to compute quantiles
of a distribution, and to generate random samples from a distribution. It is also useful for comparing and contrasting
different distributions.
```

```python
Q5: What is Bernaulli Distribution? Give an Example. What is the difference between Bernoulli
Distribution and Binomial Distribution?

Ans. Bernoulli distribution is a discrete probability distribution that describes the outcomes of a single 
binary event, where the event can have only two possible outcomes - success or failure. The distribution is named 
after the Swiss mathematician Jacob Bernoulli, who used it to model the probability of getting a head or a tail in a
coin toss. The Bernoulli distribution is characterized by a single parameter, p, which represents the probability of success.

An example of a Bernoulli distribution is the probability of a single coin flip resulting in heads. If p is the probability
of heads, then the probability of tails is 1-p. Thus, the Bernoulli distribution can be expressed as:

P(X = 1) = p
P(X = 0) = 1-p

The difference between the Bernoulli distribution and the binomial distribution is that the Bernoulli distribution models 
single trial of a binary event, whereas the binomial distribution models the number of successes in a fixed number of independent
trials of a binary event. The binomial distribution is the sum of n independent and identically distributed Bernoulli trials, 
where n is a fixed positive integer.

For example, if we want to know the probability of getting exactly two heads in three coin tosses, we would use the binomial
distribution. The probability of success (getting heads) is the same for each trial, and the trials are independent, so we can
use the binomial distribution with n = 3 and p = 0.5 to calculate the probability:

P(X = 2) = (3 choose 2) * 0.5^2 * 0.5^1 = 3/8

In summary, the Bernoulli distribution models a single trial of a binary event, while the binomial distribution models the number
of successes in a fixed number of independent trials of a binary event.
```

```python
Q7: Explain uniform Distribution with an example.
Ans. Uniform distribution is a type of probability distribution where every value in a given interval is equally 
likely to occur. In other words, the probability of obtaining any value in the interval is the same.

For example, consider a fair six-sided die. When you roll the die, each number from 1 to 6 has an equal probability of 
appearing, i.e., 1/6. This is an example of a discrete uniform distribution since the possible values are discrete (integers)
and each value has an equal probability.

Another example of uniform distribution can be a continuous variable such as the time taken by a worker to complete a task, 
where the time taken can vary continuously between a minimum and maximum value. If we assume that the worker is equally likely
to take any amount of time within this interval, then the time taken by the worker follows a continuous uniform distribution.

In summary, uniform distribution is a probability distribution where all possible values in a given interval are equally likely to 
occur. It is commonly used in statistical modeling to represent situations where there is no inherent preference for one value over
another in the given range of possible outcomes.
```

```python
Q8: What is the z score? State the importance of the z score.
Ans. A z-score (also known as a standard score) is a measure of how many standard deviations a data point is away 
from the mean of the dataset. The z-score is calculated by subtracting the mean of the dataset from the data point 
and then dividing the result by the standard deviation of the dataset. The formula for calculating the z-score is:

z = (x - μ) / σ

where x is the data point, μ is the mean of the dataset, and σ is the standard deviation of the dataset.

The z-score is important because it allows us to standardize different datasets and compare them on a common scale. By 
standardizing the data in this way, we can identify data points that are unusually high or low relative to the rest of the dataset.
Z-scores are widely used in statistical analysis, particularly in hypothesis testing and confidence interval calculations.

For example, suppose we want to compare the test scores of two different classes, where the first class has a mean score of 80 and
a standard deviation of 10, and the second class has a mean score of 75 and a standard deviation of 5. To compare the scores on a 
common scale, we can calculate the z-scores for each class. If a student in the first class scored 85 on the test, their z-score would
be (85 - 80) / 10 = 0.5, while a student in the second class who scored 80 on the test would have a z-score of (80 - 75) / 5 = 1.
In this way, we can compare the performance of the two classes despite their different means and standard deviations.
```

```python
Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.
Ans. The Central Limit Theorem (CLT) is a fundamental concept in probability theory and statistics. It states 
that the distribution of the sample means approaches a normal distribution as the sample size increases, regardless
of the shape of the population distribution. In other words, the mean of a large sample of independent and identically
distributed (iid) random variables will be approximately normally distributed, even if the original population is not 
normally distributed.

The CLT has significant implications for statistical inference, as it allows us to make generalizations about a population 
based on a sample. For example, we can use the sample mean to estimate the population mean, and use the sample standard deviation
to estimate the population standard deviation. This is the basis of hypothesis testing and confidence interval calculations.

The CLT also provides a way to understand the sampling distribution of means. The sampling distribution of means refers to the
distribution of means calculated from many different samples of the same size taken from the same population. According to the 
CLT, the sampling distribution of means will be approximately normal, regardless of the shape of the original population distribution,
provided that the sample size is sufficiently large (usually at least 30). This means that we can make probabilistic statements about
the population mean based on the distribution of sample means.

Overall, the Central Limit Theorem is a powerful tool for statistical analysis and is widely used in many areas of science, economics, 
finance, and other fields where large datasets are analyzed.
```

```python
Q10: State the assumptions of the Central Limit Theorem.
Ans. The Central Limit Theorem (CLT) relies on several assumptions in order to hold:

Independence: The samples are independent of each other. Each observation in the sample is selected randomly and does
not affect the selection of any other observation.

Identical distribution: Each observation in the sample is drawn from the same population with the same distribution.

Finite variance: The population from which the sample is drawn has a finite variance.

Sample size: The sample size is sufficiently large. While there is no specific threshold for what constitutes "sufficiently large,"
a rule of thumb is that the sample size should be at least 30.

If these assumptions hold, the Central Limit Theorem states that the distribution of the sample means will approach a normal 
distribution with a mean equal to the population mean and a standard deviation equal to the population standard deviation divided
by the square root of the sample size.
```

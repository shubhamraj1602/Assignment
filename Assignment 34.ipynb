```python
Q1: What is Estimation Statistics? Explain point estimate and interval estimate.
Ans. Estimation statistics is a branch of statistics that involves estimating unknown parameters of a population 
based on sample data. It is used to make inferences about the population based on the characteristics of a sample.

Point estimate and interval estimate are two common approaches to estimation in statistics:

Point Estimate: A point estimate is a single value that is used to estimate an unknown parameter of a population. 
It is calculated based on sample data and is used to estimate the value of the corresponding population parameter.
For example, if we want to estimate the average height of all students in a university, we can take a sample of students
and calculate the average height of those students. This average height is a point estimate of the population parameter.
Interval Estimate: An interval estimate, on the other hand, is a range of values that is used to estimate an unknown population
parameter. It is calculated based on sample data and takes into account the variability and uncertainty of the estimation process.
An interval estimate is typically expressed in the form of a confidence interval, which provides a range of values within which the
true population parameter is likely to lie. For example, if we want to estimate the average height of all students in a university,
we can calculate a confidence interval based on a sample of students. This interval estimate provides a range of values within which
the true population parameter is likely to lie with a certain level of confidence.
```

```python
Q2. Write a Python function to estimate the population mean using a sample mean and standard deviation.
Ans. def population_mean(sample_mean, sample_std, sample_size):
    """
    Calculates the estimated population mean based on a sample mean, sample standard deviation, and sample size.
    
    Args:
    - sample_mean: float, the mean of the sample
    - sample_std: float, the standard deviation of the sample
    - sample_size: int, the size of the sample
    
    Returns:
    - float, the estimated population mean
    """
    import math
    std_error = sample_std / math.sqrt(sample_size)  # standard error
    z_score = 1.96  # assuming 95% confidence level
    margin_error = z_score * std_error
    pop_mean = sample_mean - margin_error  # lower bound of the confidence interval
    return pop_mean
```

```python
Q3: What is Hypothesis testing? Why is it used? State the importance of Hypothesis testing.
Ans. Hypothesis testing is a statistical method used to make decisions or draw conclusions about a population based 
on sample data. It involves two hypotheses, the null hypothesis and the alternative hypothesis, which are both statements 
about the population parameter being tested.

The null hypothesis typically states that there is no difference or no effect between the sample and the population, while
the alternative hypothesis suggests otherwise. Hypothesis testing is used to determine the likelihood of observing a sample
statistic if the null hypothesis is true, and to make a decision whether to reject or fail to reject the null hypothesis based 
on the level of significance.

The importance of hypothesis testing lies in its ability to help us make decisions and draw conclusions based on data. By using
hypothesis testing, we can determine the validity of assumptions, test the effectiveness of treatments or interventions, and
make decisions about policies or procedures.

For example, in a clinical trial, hypothesis testing can be used to determine whether a new drug is effective in treating a
particular disease. The null hypothesis would be that the drug has no effect, while the alternative hypothesis would be that
it does have an effect. By conducting the trial and analyzing the results, we can determine the probability of observing the
results if the null hypothesis is true and decide whether to reject or fail to reject it.

In summary, hypothesis testing is a powerful tool for making decisions and drawing conclusions based on data, and it is widely
used in many fields such as science, medicine, economics, and psychology.
```

```python
Q6: What is a null and alternative hypothesis? Give some examples.
Ans. A null hypothesis is a statement that assumes there is no significant difference between two groups being 
compared or no significant effect of a treatment on a population. It is denoted by H0.

An alternative hypothesis, on the other hand, is a statement that contradicts the null hypothesis, suggesting that
there is a significant difference between the groups or a significant effect of the treatment on the population. It is
denoted by H1 or Ha.

Here are some examples of null and alternative hypotheses:

Null Hypothesis: There is no difference in the mean heights of men and women in the population.
Alternative Hypothesis: The mean height of men is greater than the mean height of women in the population.

Null Hypothesis: There is no relationship between smoking and lung cancer.
Alternative Hypothesis: Smoking is associated with an increased risk of lung cancer.

Null Hypothesis: There is no difference in the effectiveness of two medications.
Alternative Hypothesis: One medication is more effective than the other.

Null Hypothesis: There is no difference in the mean test scores of two groups of students.
Alternative Hypothesis: The mean test score of one group of students is greater than the other.
```

```python
Q7: Write down the steps involved in hypothesis testing.
Ans. The steps involved in hypothesis testing are:

State the hypothesis: The first step in hypothesis testing is to clearly state the null and alternative hypotheses.

Set the significance level: The significance level, also known as the alpha level, is the level of risk that the researcher is willing 
to take in rejecting the null hypothesis when it is actually true. A common significance level is 0.05.

Determine the test statistic: The test statistic is a value that is calculated from the sample data and is used to determine the p-value.

Calculate the p-value: The p-value is the probability of obtaining a test statistic as extreme as, or more extreme than, the 
observed test statistic, assuming that the null hypothesis is true.

Make a decision: Based on the p-value and the significance level, the researcher can make a decision whether to reject or fail 
to reject the null hypothesis.

Interpret the results: If the null hypothesis is rejected, it means that there is enough evidence to support the alternative hypothesis
. If the null hypothesis is not rejected, it means that there is not enough evidence to support the alternative hypothesis.

Draw conclusions: The final step is to draw conclusions based on the results of the hypothesis test and to communicate the 
findings to the relevant stakeholders.
```

```python
Q8. Define p-value and explain its significance in hypothesis testing.
Ans. The p-value is the probability of obtaining a test statistic at least as extreme as the one observed, assuming 
the null hypothesis is true. In other words, it represents the probability of observing the data or a more extreme 
result if the null hypothesis is actually true.

The significance of the p-value lies in its use in hypothesis testing. After conducting a hypothesis test, if the p-value
is less than or equal to the level of significance (usually 0.05), then we reject the null hypothesis in favor of the alternative
hypothesis. A smaller p-value indicates stronger evidence against the null hypothesis and provides support for the alternative hypothesis.
On the other hand, a larger p-value suggests weak evidence against the null hypothesis and indicates that we should not reject the
null hypothesis.

Therefore, the p-value plays a critical role in hypothesis testing as it helps us make decisions about the statistical significance
of our results and provides a measure of the strength of evidence against the null hypothesis.
```

```python
Q10. Write a Python program to calculate the two-sample t-test for independent samples, given two
random samples of equal size and a null hypothesis that the population means are equal.
```

```python
import numpy as np
from scipy import stats

# Enter the data for two samples
sample1 = [4, 6, 8, 10, 12]
sample2 = [5, 7, 9, 11, 13]

# Calculate the means and standard deviations of the two samples
mean1 = np.mean(sample1)
mean2 = np.mean(sample2)
std1 = np.std(sample1, ddof=1)
std2 = np.std(sample2, ddof=1)

# Calculate the t-statistic and degrees of freedom
t = (mean1 - mean2) / np.sqrt((std1**2 / len(sample1)) + (std2**2 / len(sample2)))
df = len(sample1) + len(sample2) - 2

# Calculate the p-value
p_value = stats.t.sf(np.abs(t), df) * 2

# Print the results
print("t-statistic: ", t)
print("p-value: ", p_value)
```

```python
Q12: What is t-statistic? State the formula for t-statistic.
Ans. The t-statistic is a test statistic used in hypothesis testing when the sample size is small and the population
standard deviation is unknown. It is used to test the difference between the means of two independent samples.

The formula for t-statistic is:

t = (x̄1 - x̄2) / (s_p * sqrt(1/n1 + 1/n2))

where x̄1 and x̄2 are the sample means of the two samples, s_p is the pooled standard deviation, n1 and n2 are the sample 
sizes of the two samples.
```

```python
Q13. A coffee shop owner wants to estimate the average daily revenue for their shop. They take a random
sample of 50 days and find the sample mean revenue to be $500 with a standard deviation of $50.
Estimate the population mean revenue with a 95% confidence interval.

Ans. To estimate the population mean revenue with a 95% confidence interval, we can use the following formula:

Confidence interval = sample mean ± (t-value x standard error)

where the t-value is obtained from the t-distribution table for the given degrees of freedom and desired level of confidence,
and the standard error is calculated as:

Standard error = sample standard deviation / sqrt(sample size)

Here, we have a sample size of 50, a sample mean of $500, and a sample standard deviation of $50. We want to calculate the 95% 
confidence interval, which corresponds to a level of significance of 0.05, and we have 49 degrees of freedom (50 - 1).

From the t-distribution table, the t-value for a 95% confidence interval and 49 degrees of freedom is approximately 2.009.

Substituting the values into the formula, we get:
    
Confidence interval = $500 ± (2.009 x ($50 / sqrt(50)))
= $500 ± $14.17
= ($485.83, $514.17)

Therefore, we can estimate with 95% confidence that the population mean revenue is between $485.83 and $514.17.
```

```python
Q14. A researcher hypothesizes that a new drug will decrease blood pressure by 10 mmHg. They conduct a
clinical trial with 100 patients and find that the sample mean decrease in blood pressure is 8 mmHg with a
standard deviation of 3 mmHg. Test the hypothesis with a significance level of 0.05.

Ans.To test the hypothesis, we need to perform a one-sample t-test with the following null and alternative hypotheses:

Null hypothesis: The true mean decrease in blood pressure is equal to 10 mmHg.
Alternative hypothesis: The true mean decrease in blood pressure is less than 10 mmHg.
We can calculate the t-statistic using the formula:

t = (sample mean - hypothesized mean) / (standard error)

where the standard error is calculated as:

SE = standard deviation / sqrt(sample size)

Plugging in the given values, we get:

t = (8 - 10) / (3 / sqrt(100))
t = -4.47

Using a t-table or calculator, we find the p-value associated with this t-value and degrees of freedom (df = n - 1 = 99) 
to be very small, much less than 0.05. Therefore, we reject the null hypothesis and conclude that the new drug does decrease
blood pressure by a statistically significant amount.
```

```python
Q15. An electronics company produces a certain type of product with a mean weight of 5 pounds and a
standard deviation of 0.5 pounds. A random sample of 25 products is taken, and the sample mean weight
is found to be 4.8 pounds. Test the hypothesis that the true mean weight of the products is less than 5
pounds with a significance level of 0.01.

Ans. To test the hypothesis, we will use a one-tailed t-test with a significance level of 0.01.

Null hypothesis: The true mean weight of the products is equal to 5 pounds.

Alternative hypothesis: The true mean weight of the products is less than 5 pounds.

The test statistic can be calculated using the formula:

t = (x̄ - μ) / (s / sqrt(n))

where x̄ is the sample mean, μ is the population mean (5 pounds), s is the sample standard deviation (0.5 pounds), and n is 
the sample size (25).

Substituting the values, we get:

t = (4.8 - 5) / (0.5 / sqrt(25))
t = -2

Using a t-distribution table with 24 degrees of freedom (n-1), the critical value for a one-tailed test with a significance 
level of 0.01 is -2.492.

Since the calculated t-value (-2) is less than the critical value (-2.492), we reject the null hypothesis. Therefore, we can
conclude that the true mean weight of the products is less than 5 pounds with a significance level of 0.01.
```

```python
Q16. Two groups of students are given different study materials to prepare for a test. The first group (n1 =
30) has a mean score of 80 with a standard deviation of 10, and the second group (n2 = 40) has a mean
score of 75 with a standard deviation of 8. Test the hypothesis that the population means for the two
groups are equal with a significance level of 0.01.

Ans. To test the hypothesis that the population means for the two groups are equal, we can use a two-sample t-test. The
null hypothesis is that the population means are equal, and the alternative hypothesis is that they are not equal.

The formula for the two-sample t-test is:

t = (x1 - x2) / sqrt((s1^2 / n1) + (s2^2 / n2))

Where:

x1 and x2 are the sample means for the two groups
s1 and s2 are the sample standard deviations for the two groups
n1 and n2 are the sample sizes for the two groups

With a significance level of 0.01 and two-tailed test, the critical t-value is ±2.704, which can be obtained from a t-table or calculator.

Now we can calculate the t-statistic:

t = (80 - 75) / sqrt((10^2 / 30) + (8^2 / 40))
t = 2.44

Since the calculated t-value of 2.44 is greater than the critical t-value of 2.704, we reject the null hypothesis and conclude
that the population means for the two groups are not equal at the 0.01 significance level. This suggests that the study materials
given to the two groups had a different effect on their test scores.
```

```python
Q17. A marketing company wants to estimate the average number of ads watched by viewers during a TV
program. They take a random sample of 50 viewers and find that the sample mean is 4 with a standard
deviation of 1.5. Estimate the population mean with a 99% confidence interval.

Ans. To estimate the population mean with a 99% confidence interval, we can use the formula:

Confidence interval = sample mean ± (critical value) x (standard error)

where the critical value is obtained from the t-distribution table based on the sample size and the desired confidence level, 
and the standard error is calculated as the sample standard deviation divided by the square root of the sample size.

The sample size is 50, and the sample mean is 4 with a standard deviation of 1.5. The degrees of freedom for a 99% confidence 
interval are 49 (n-1). From the t-distribution table, the critical value for a 99% confidence interval with 49 degrees of freedom is 2.680.

The standard error can be calculated as:

standard error = sample standard deviation / sqrt(sample size)
standard error = 1.5 / sqrt(50)
standard error = 0.212

Substituting the values in the formula, we get:
    
Confidence interval = 4 ± 2.680 x 0.212
Confidence interval = 4 ± 0.569

Therefore, the 99% confidence interval estimate for the population mean number of ads watched by viewers during a 
TV program is between 3.431 and 4.569. We can be 99% confident that the true population mean lies within this interval.
```

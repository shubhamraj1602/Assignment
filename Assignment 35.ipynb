```python
Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would
use each type of test.

Ans. A t-test and a z-test are both statistical tests used to make inferences about population parameters based on sample
statistics, but they differ in the conditions under which they are applicable.

A z-test is used when the sample size is large (usually n > 30), and the population standard deviation is known or can be
assumed to be known. The test statistic is the z-score, which measures the number of standard deviations between the sample
mean and the population mean. Z-tests are often used in quality control and manufacturing settings, where large sample sizes are common.

For example, suppose a shoe factory wants to test whether the mean length of a batch of shoes is equal to a target value of
10 inches. The factory takes a random sample of 100 shoes and finds that the sample mean length is 10.2 inches, with a known 
population standard deviation of 0.5 inches. A z-test can be used to determine whether the sample mean length is significantly
different from the target value of 10 inches.

On the other hand, a t-test is used when the sample size is small (usually n < 30) or when the population standard deviation is
unknown. The test statistic is the t-score, which is calculated using the sample mean, the sample standard deviation, and the
degrees of freedom (n-1). T-tests are commonly used in social sciences, psychology, and medicine, where sample sizes tend to be smaller.

For example, suppose a researcher wants to test whether a new medication reduces blood pressure in patients. The researcher
takes a random sample of 20 patients and measures their blood pressure before and after taking the medication. The difference
between the two measurements is the sample mean, and the variability in the differences is the sample standard deviation. Since
the population standard deviation is unknown, a t-test can be used to determine whether the mean difference in blood pressure is 
significantly different from zero.

In summary, a z-test is used when the sample size is large and the population standard deviation is known or can be assumed to be
known, while a t-test is used when the sample size is small or the population standard deviation is unknown.
```

```python
Q2: Differentiate between one-tailed and two-tailed tests.
Ans. One-tailed and two-tailed tests are two types of hypothesis tests used in statistics.

A one-tailed test is a hypothesis test in which the alternative hypothesis is directional, meaning that it specifies the
direction of the difference or effect that is being tested. In a one-tailed test, the null hypothesis assumes that there is
no difference or effect, while the alternative hypothesis specifies the direction of the difference or effect. The one-tailed
test is used when the researcher has a specific directional prediction about the relationship between the variables being tested.

For example, suppose a researcher wants to test whether a new drug is more effective than a placebo in treating a particular disease.
The researcher's directional prediction is that the new drug will be more effective than the placebo, so the alternative hypothesis
would state that the mean difference between the two treatments is greater than zero. A one-tailed test would be used to test this
hypothesis.

A two-tailed test, on the other hand, is a hypothesis test in which the alternative hypothesis is non-directional, meaning that it
does not specify the direction of the difference or effect that is being tested. In a two-tailed test, the null hypothesis assumes
that there is no difference or effect, while the alternative hypothesis states that there is a difference or effect, but it does not
specify the direction. The two-tailed test is used when the researcher does not have a specific directional prediction about the 
relationship between the variables being tested.

For example, suppose a researcher wants to test whether there is a difference in IQ scores between two groups of students. The
researcher does not have a directional prediction about the relationship between the groups, so the alternative hypothesis would 
simply state that the mean difference between the groups is not equal to zero. A two-tailed test would be used to test this hypothesis.

In summary, a one-tailed test is used when the researcher has a specific directional prediction about the relationship between the
variables being tested, while a two-tailed test is used when the researcher does not have a specific directional prediction.
```

```python
Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for
each type of error.

Ans. In hypothesis testing, there are two types of errors that can occur: Type 1 error and Type 2 error.

Type 1 error occurs when the null hypothesis is rejected, even though it is actually true. In other words, Type 1 error 
occurs when we conclude that there is a significant effect or difference between two groups when in reality there is none. 
The probability of making a Type 1 error is denoted by the Greek letter alpha (α), and it is typically set at 0.05 or 0.01 
in scientific research.

An example scenario of Type 1 error is a clinical trial where a new drug is being tested to see if it is effective in treating a
certain disease. The null hypothesis is that the drug has no effect, and the alternative hypothesis is that the drug is effective.
If the researchers reject the null hypothesis and conclude that the drug is effective, when in fact it is not, then they have made 
a Type 1 error. This means that they will mistakenly conclude that the drug works and may prescribe it to patients, even though it
does not actually help them.

Type 2 error occurs when the null hypothesis is not rejected, even though it is actually false. In other words, Type 2 error occurs
when we conclude that there is no significant effect or difference between two groups when in reality there is one. The probability
of making a Type 2 error is denoted by the Greek letter beta (β), and it is affected by the sample size, effect size, and the significance
level of the test.

An example scenario of Type 2 error is a criminal trial where the defendant is accused of committing a crime. The null hypothesis is that 
the defendant is innocent, and the alternative hypothesis is that the defendant is guilty. If the jury fails to reject the null hypothesis
and concludes that the defendant is innocent, even though he or she is actually guilty, then they have made a Type 2 error. This means
that the defendant will be acquitted, even though he or she is guilty of the crime.

In summary, Type 1 error occurs when we wrongly reject a true null hypothesis, and Type 2 error occurs when we wrongly accept a false
null hypothesis. Both types of errors can have serious consequences in scientific research and decision-making, and it is important to
control the probabilities of making these errors in hypothesis testing.
```

```python
Q4: Explain Bayes's theorem with an example.
Ans. Bayes's theorem is a mathematical formula used in probability theory to calculate the probability of a 
hypothesis (or event) given the available evidence. The theorem is named after the 18th-century English mathematician
and theologian, Thomas Bayes.

The formula for Bayes's theorem is:

P(H|E) = P(E|H) * P(H) / P(E)

where:

P(H|E) is the probability of the hypothesis H given the evidence E
P(E|H) is the probability of the evidence E given the hypothesis H
P(H) is the prior probability of the hypothesis H
P(E) is the prior probability of the evidence E
To understand Bayes's theorem, let's consider an example:

Suppose a woman wants to know the probability of having breast cancer given that she has a positive mammogram result. She 
consults a doctor who tells her that, in general, 1% of women over 50 have breast cancer. The mammogram is 90% accurate in 
detecting breast cancer in women who actually have it and also 90% accurate in giving a negative result in women who do not have cancer.

Using Bayes's theorem, we can calculate the probability of having breast cancer given a positive mammogram result as follows:

Let H be the hypothesis that the woman has breast cancer, and let E be the evidence of a positive mammogram result.
P(H) = 0.01 (prior probability of having breast cancer)
P(E|H) = 0.9 (probability of a positive mammogram given the woman has breast cancer)
P(not H) = 0.99 (prior probability of not having breast cancer)
P(E|not H) = 0.1 (probability of a positive mammogram given the woman does not have breast cancer)
Using these values, we can calculate the probability of having breast cancer given a positive mammogram result as:

P(H|E) = P(E|H) * P(H) / [P(E|H) * P(H) + P(E|not H) * P(not H)]
= 0.9 * 0.01 / [0.9 * 0.01 + 0.1 * 0.99]
= 0.0833

Therefore, the probability of the woman having breast cancer given a positive mammogram result is about 8.33%. This means 
that even though she has a positive mammogram result, there is still a relatively low probability of her having breast cancer.

In summary, Bayes's theorem is a powerful tool for updating the probability of a hypothesis or event based on new evidence or 
information. It is widely used in fields such as medicine, finance, and machine learning to make informed decisions and predictions.
```

```python
Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.
Ans. A confidence interval is a range of values that is likely to contain the true value of a population parameter
with a certain degree of confidence or probability. It is commonly used in statistical inference to estimate the population
parameter based on a sample of data.

To calculate a confidence interval, we need to specify a confidence level, which is the probability that the interval contains
the true population parameter. A common confidence level is 95%, which means that if we repeated the sampling process many times
and calculated the confidence interval each time, about 95% of the intervals would contain the true population parameter.

The formula for a confidence interval for a population mean with known population standard deviation is:

Confidence Interval = sample mean +/- z-score * standard error

where:

sample mean is the mean of the sample data
z-score is the critical value from the standard normal distribution based on the desired confidence level
standard error is the standard deviation of the sampling distribution, which is equal to the population standard deviation divided by
the square root of the sample size
For example, suppose we want to estimate the average height of male adults in a population. We take a random sample of 50 male adults
and find that the sample mean height is 175 cm, and the population standard deviation is 5 cm. We want to calculate a 95% confidence
interval for the population mean height.

The critical value for a 95% confidence interval from the standard normal distribution is 1.96. The standard error is calculated as:

standard error = population standard deviation / sqrt(sample size)
= 5 / sqrt(50)
= 0.707

Therefore, the 95% confidence interval for the population mean height is:

175 +/- 1.96 * 0.707 = (173.62, 176.38)

This means that we are 95% confident that the true population mean height of male adults lies between 173.62 cm and 176.38 cm. Note 
that if we repeat the sampling process many times, about 95% of the intervals we calculate would contain the true population mean height.
```

```python
Q6. Use Bayes' Theorem to calculate the probability of an event occurring given prior knowledge of the
event's probability and new evidence. Provide a sample problem and solution.

Ans. Sure, here's an example problem to demonstrate the use of Bayes' Theorem:

Suppose a medical test for a certain disease has a 95% accuracy rate, meaning that the probability of a true positive or 
true negative result is 0.95, and the probability of a false positive or false negative result is 0.05. The prevalence of 
the disease in the population is 1%. If a person tests positive for the disease, what is the probability that they actually
have the disease?

Solution:

Let A be the event that the person has the disease, and B be the event that the person tests positive. We want to calculate the
conditional probability of A given B, denoted as P(A|B), using Bayes' Theorem:

P(A|B) = P(B|A) * P(A) / P(B)

where:

P(B|A) is the probability of testing positive given that the person has the disease, which is the accuracy rate of the test: P(B|A) = 0.95
P(A) is the prior probability of having the disease, which is given as 0.01
P(B) is the marginal probability of testing positive, which is the probability of testing positive given that the person has the
disease (true positive) plus the probability of testing positive given that the person does not have the disease (false positive): 
    P(B) = P(B|A) * P(A) + P(B|not A) * P(not A)
P(B|not A) is the probability of testing positive given that the person does not have the disease, which is the false positive rate:
    P(B|not A) = 0.05
P(not A) is the complement of P(A), which is 1 - P(A) = 0.99
Substituting the given values into the formula, we have:

P(A|B) = 0.95 * 0.01 / (0.95 * 0.01 + 0.05 * 0.99)
= 0.161

Therefore, the probability that a person who tests positive for the disease actually has the disease is about 16.1%. This means that
even if a person tests positive for the disease, there is still an 83.9% chance that they do not have the disease.
```

```python
Q8. What is the margin of error in a confidence interval? How does sample size affect the margin of error?
Provide an example of a scenario where a larger sample size would result in a smaller margin of error.

Ans. The margin of error is a measure of the uncertainty or precision associated with a sample estimate, such as a sample
mean or proportion, when we use it to make inferences about the population parameter. It represents the range of values
that we can be reasonably confident includes the true population parameter. The larger the margin of error, the less precise
our estimate is.

The margin of error is calculated as the product of the critical value (which depends on the desired confidence level) and the
standard error (which is a function of the sample size and the population standard deviation or the sample standard deviation).

The sample size affects the margin of error in the following way: larger sample sizes tend to have smaller margins of error, while
smaller sample sizes tend to have larger margins of error. This is because as the sample size increases, the standard error decreases,
which in turn reduces the margin of error.

For example, suppose we want to estimate the proportion of registered voters in a city who support a certain ballot measure, and we 
take a sample of 200 voters. If we calculate a 95% confidence interval using the sample proportion and assume that the population
standard deviation is known to be 0.5, we might get a margin of error of 0.07. This means that we can be 95% confident that the true 
proportion of voters who support the ballot measure falls within the range of the sample proportion plus or minus 0.07.

Now, suppose we increase the sample size to 1000 voters while keeping all other factors the same. The standard error will decrease
because the sample size is larger, and so will the margin of error. Assuming that the population standard deviation is still 0.5, we 
might get a margin of error of 0.022. This means that we can be more confident in our estimate with a larger sample size.

In summary, the margin of error measures the precision of our estimate and is influenced by the sample size. A larger sample size will
generally result in a smaller margin of error.
```

```python
Q9. Calculate the z-score for a data point with a value of 75, a population mean of 70, and a population
standard deviation of 5. Interpret the results.

Ans. To calculate the z-score, we use the formula:

z = (x - μ) / σ

Where:
x = the data point value
μ = the population mean
σ = the population standard deviation

Plugging in the values given in the question, we get:

z = (75 - 70) / 5 = 1

Interpretation: The z-score of 1 means that the data point of 75 is one standard deviation above the population mean of 70.
This suggests that the data point is relatively high compared to the rest of the population. A z-score of 1 also indicates 
that the data point is within the first quartile of the population distribution.
```

```python
Q10. In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average
of 6 pounds with a standard deviation of 2.5 pounds. Conduct a hypothesis test to determine if the drug is
significantly effective at a 95% confidence level using a t-test.

Ans. To test whether the weight loss drug is significantly effective, we can use a one-sample t-test.

The null hypothesis (H0) is that the mean weight loss is not significantly different from zero (μ = 0), and the 
alternative hypothesis (Ha) is that the mean weight loss is significantly different from zero (μ ≠ 0).

We can calculate the t-statistic using the formula:

t = (x̄ - μ) / (s / sqrt(n))

where:

x̄ is the sample mean (6 pounds in this case)
μ is the hypothesized population mean (0 pounds in this case)
s is the sample standard deviation (2.5 pounds in this case)
n is the sample size (50 in this case)
Substituting the values we get:

t = (6 - 0) / (2.5 / sqrt(50))

t = 12

Using a t-table or calculator, with 49 degrees of freedom (df = n - 1), we can find that the critical t-value at a 95% 
confidence level (two-tailed test) is ±2.009.

Since the calculated t-value (12) is greater than the critical t-value (2.009), we reject the null hypothesis and conclude
that the weight loss drug is significantly effective at a 95% confidence level.
```

```python
Q13. A population has a mean of 60 and a standard deviation of 8. A sample of 50 observations has a mean
of 65. Calculate the 90% confidence interval for the true population mean.

Ans. To calculate the 90% confidence interval for the true population mean, we can use the following formula:

CI = x̄ ± z*(s/√n)

where:

CI is the confidence interval
x̄ is the sample mean (65 in this case)
z is the critical value from the standard normal distribution for a 90% confidence level (which is 1.645)
s is the population standard deviation (8 in this case)
n is the sample size (50)
Substituting the values we get:

CI = 65 ± 1.645*(8/√50)

CI = 65 ± 1.81

CI = (63.19, 66.81)

Therefore, we can say with 90% confidence that the true population mean is between 63.19 and 66.81.
```

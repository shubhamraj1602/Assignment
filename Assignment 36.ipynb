```python
Q1. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation
of 5 using Python. Interpret the results.
```

```python
import scipy.stats as stats

sample_mean = 50
sample_std = 5
sample_size = 100 # assuming a sample size of 100 for demonstration purposes

confidence_level = 0.95

# calculate the margin of error
margin_of_error = stats.norm.ppf((1 + confidence_level) / 2) * (sample_std / (sample_size ** 0.5))

# calculate the lower and upper bounds of the confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

# print the results
print("95% confidence interval: [{:.2f}, {:.2f}]".format(lower_bound, upper_bound))
```

```python
Q2. Conduct a chi-square goodness of fit test to determine if the distribution of colors of M&Ms in a bag
matches the expected distribution of 20% blue, 20% orange, 20% green, 10% yellow, 10% red, and 20%
brown. Use Python to perform the test with a significance level of 0.05.
```

```python
import numpy as np
import scipy.stats as stats

# observed frequencies of colors in a sample of M&Ms
observed_freq = [25, 17, 19, 12, 10, 17]  # blue, orange, green, yellow, red, brown

# expected frequencies of colors based on the given distribution
expected_freq = [0.2, 0.2, 0.2, 0.1, 0.1, 0.2] * sum(observed_freq)

# calculate the chi-square test statistic and p-value
chi2, p_value = stats.chisquare(observed_freq, expected_freq)

# calculate the degrees of freedom
df = len(observed_freq) - 1

# calculate the critical chi-square value
alpha = 0.05
critical_chi2 = stats.chi2.ppf(q=1-alpha, df=df)

# print the results
print("Chi-square test statistic:", chi2)
print("P-value:", p_value)
print("Critical chi-square value:", critical_chi2)

# check if the null hypothesis is rejected or not
if chi2 > critical_chi2:
    print("Reject the null hypothesis: The distribution of colors of M&Ms in the bag does not match the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The distribution of colors of M&Ms in the bag matches the expected distribution.")
```

```python
Q5. Calculate the 90% confidence interval for a sample of data with a mean of 75 and a standard deviation
of 12 using Python. Interpret the results.
```

```python
import scipy.stats as stats

# set the sample size
n = 1

# set the sample mean and standard deviation
sample_mean = 75
sample_std = 12

# set the confidence level
conf_level = 0.9

# calculate the margin of error
margin_error = stats.norm.ppf((1 + conf_level) / 2) * (sample_std / n**0.5)

# calculate the confidence interval
conf_int = (sample_mean - margin_error, sample_mean + margin_error)

# print the results
print(f"The 90% confidence interval for the sample mean is {conf_int}")
```

```python
Q6. Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the
area corresponding to a chi-square statistic of 15.
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# set the degrees of freedom
df = 10

# set the x-axis range
x = np.linspace(0, 30, 500)

# plot the chi-square distribution
plt.plot(x, stats.chi2.pdf(x, df))

# shade the area corresponding to a chi-square statistic of 15
x_shade = np.linspace(15, 30, 100)
y_shade = stats.chi2.pdf(x_shade, df)
plt.fill_between(x_shade, y_shade, color='blue', alpha=0.3)

# label the axes
plt.xlabel("Chi-Square Statistic")
plt.ylabel("Probability Density")

# show the plot
plt.show()
```

```python
Q7. A random sample of 1000 people was asked if they preferred Coke or Pepsi. Of the sample, 520
preferred Coke. Calculate a 99% confidence interval for the true proportion of people in the population who
prefer Coke.

Ans. To calculate the 99% confidence interval for the true proportion of people in the population who prefer Coke, we can use the formula:

CI = p̂ ± z*√(p̂(1-p̂)/n)

where:
p̂ = proportion of people in the sample who prefer Coke
n = sample size
z = critical value from the standard normal distribution for the desired confidence level (99% in this case)

First, let's calculate the sample proportion:

p̂ = 520/1000 = 0.52

Next, let's find the critical value from the standard normal distribution:

z = stats.norm.ppf(0.995) # 99% confidence level, two-tailed test

z = 2.5758293035489004

Now, we can calculate the confidence interval:

CI = 0.52 ± 2.5758√(0.52(1-0.52)/1000)

CI = (0.4859, 0.5541)

Therefore, we can say with 99% confidence that the true proportion of people in the population who prefer Coke is between 0.4859 and 0.5541.
```

```python
Q8. A researcher hypothesizes that a coin is biased towards tails. They flip the coin 100 times and observe
45 tails. Conduct a chi-square goodness of fit test to determine if the observed frequencies match the
expected frequencies of a fair coin. Use a significance level of 0.05.

Ans. To conduct a chi-square goodness of fit test, we need to follow these steps:

Step 1: State the null hypothesis and the alternative hypothesis:

Null hypothesis (H0): The observed frequencies match the expected frequencies of a fair coin.
Alternative hypothesis (Ha): The observed frequencies do not match the expected frequencies of a fair coin.

Step 2: Determine the level of significance and the degrees of freedom:

Level of significance: α = 0.05
Degrees of freedom: df = k - 1, where k is the number of categories (in this case, k = 2: heads and tails)

Step 3: Define the expected frequencies:

Under the assumption of a fair coin, the expected frequency for each outcome (heads and tails) is 50.

Step 4: Calculate the test statistic:

The chi-square test statistic for a goodness of fit test is calculated as follows:

χ2 = ∑((Oi - Ei)^2 / Ei), where Oi is the observed frequency, Ei is the expected frequency, and the summation is taken over all categories.

Using the given data, we have:

Oi (observed frequency of tails) = 45
Oi (observed frequency of heads) = 100 - 45 = 55
Ei (expected frequency of tails) = 50
Ei (expected frequency of heads) = 50

So, the chi-square test statistic is:

χ2 = ((45-50)^2/50) + ((55-50)^2/50) = 0.5

Step 5: Determine the critical value:

We use the chi-square distribution table with df = 1 and α = 0.05 to find the critical value of χ2. The critical value is 3.84.

Step 6: Make a decision:

Since the calculated test statistic (0.5) is less than the critical value (3.84), we fail to reject the null hypothesis. 
This means that there is not enough evidence to conclude that the coin is biased towards tails. Therefore, we accept the null
hypothesis and conclude that the observed frequencies match the expected frequencies of a fair coin at the significance level of 0.05.
```

```python
Q9. A study was conducted to determine if there is an association between smoking status (smoker or
non-smoker) and lung cancer diagnosis (yes or no). The results are shown in the contingency table below.
Conduct a chi-square test for independence to determine if there is a significant association between
smoking status and lung cancer diagnosis.

       Lung Cancer: Yes      Lung Cancer: No

Smoker       60                  140
Non-smoker   30                  170

Use a significance level of 0.05.

Ans. To conduct a chi-square test for independence to determine if there is a significant association between smoking 
status and lung cancer diagnosis, we can use the contingency table provided.

The null hypothesis (H0) is that there is no association between smoking status and lung cancer diagnosis, and the alternative
hypothesis (Ha) is that there is a significant association between the two variables.

H0: The variables are independent.
Ha: The variables are not independent.

We can use the chi-square statistic formula to calculate the test statistic:

χ2 = Σ (Oi - Ei)^2 / Ei

where Oi is the observed frequency in each cell, Ei is the expected frequency in each cell under the assumption of independence,
and Σ is the sum of all cells.

We can calculate the expected frequencies by taking the row and column totals and dividing by the total sample size:
    
                  Lung Cancer: Yes         Lung Cancer: No

Smoker 60/330 = 0.1818 140/330 = 0.4242 1
Non-smoker 30/330 = 0.0909 170/330 = 0.5152 1
Total 0.2727 0.9394 2

We can then calculate the expected frequencies for each cell by multiplying the row total by the column total and dividing by the total sample size:

              Lung Cancer: Yes         Lung Cancer: No
        
Smoker 0.1818 * 0.2727 = 0.0495 0.4242 * 0.2727 = 0.1156
Non-smoker 0.0909 * 0.2727 = 0.0247 0.5152 * 0.2727 = 0.1406

Using these expected frequencies, we can calculate the chi-square test statistic:

χ2 = [(60-0.0495)^2/0.0495 + (140-0.1156)^2/0.1156 + (30-0.0247)^2/0.0247 + (170-0.1406)^2/0.1406]

χ2 = 28.89

Using a chi-square distribution table with (2-1)*(2-1) = 1 degree of freedom and a significance level of 0.05, the critical value is 3.84.

Since the calculated chi-square value (28.89) is greater than the critical value (3.84), we reject the null hypothesis.

Therefore, we can conclude that there is a significant association between smoking status and lung cancer diagnosis at a 
significance level of 0.05.
```

```python
Q11. A random sample of 30 people was selected from a population with an unknown mean and standard
deviation. The sample mean was found to be 72 and the sample standard deviation was found to be 10.
Conduct a hypothesis test to determine if the population mean is significantly different from 70. Use a
significance level of 0.05.

ANs. To conduct a hypothesis test to determine if the population mean is significantly different from 70, we can use 
a two-tailed t-test with a significance level of 0.05.

The null hypothesis (H0) is that the population mean is equal to 70, and the alternative hypothesis (Ha) is that
the population mean is significantly different from 70.

H0: μ = 70
Ha: μ ≠ 70

We can use the t-statistic formula to calculate the test statistic:

t = (x̄ - μ) / (s / √n)

where x̄ is the sample mean, μ is the population mean, s is the sample standard deviation, and n is the sample size.

Substituting the values we get:

t = (72 - 70) / (10 / √30) = 1.3856

The degrees of freedom for the t-distribution are n - 1 = 29.

Using a two-tailed t-test with a significance level of 0.05 and 29 degrees of freedom, the critical t-value is ±2.045.

Since the calculated t-value (1.3856) is within the range of the critical t-values (-2.045 to 2.045), we fail to reject the null hypothesis.

Therefore, we can conclude that there is not enough evidence to suggest that the population mean is significantly different
from 70 at a significance level of 0.05.
```

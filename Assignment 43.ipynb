```python
Q1. What is data encoding? How is it useful in data science?
Ans. Data encoding is the process of converting data from one format or representation to another format, such as
transforming categorical data into numerical data. Data encoding is useful in data science because it allows data
scientists to analyze data more effectively by making it suitable for machine learning algorithms, data visualization,
and statistical analysis. It can help in improving the accuracy of models, reduce the dimensionality of the dataset, 
and prevent overfitting.

Q2. What is nominal encoding? Provide an example of how you would use it in a real-world scenario.
Ams. Nominal encoding is a technique used to transform categorical data into numerical data by assigning a unique integer 
value to each category. For instance, consider a dataset that contains information about customer reviews of a restaurant.
One of the columns in the dataset could be the rating of the restaurant, which could take on the values "Excellent,"
"Good," "Average," "Poor," and "Terrible." Nominal encoding could be used to transform this column into numerical data,
where "Excellent" is assigned a value of 1, "Good" is assigned a value of 2, "Average" is assigned a value of 3, "Poor" 
is assigned a value of 4, and "Terrible" is assigned a value of 5.

Q3. In what situations is nominal encoding preferred over one-hot encoding? Provide a practical example.
Ans. Nominal encoding is preferred over one-hot encoding when the categorical variable has a large number of categories,
and the categories are unordered. For instance, consider a dataset that contains information about customer preferences 
for different types of food. One of the columns in the dataset could be the type of cuisine, which could take on the values
"Chinese," "Italian," "Mexican," "Japanese," and "Indian." Nominal encoding would be preferred over one-hot encoding in
this case because the categories are unordered, and the number of categories is relatively small.

Q4. Suppose you have a dataset containing categorical data with 5 unique values. Which encoding technique would you use 
to transform this data into a format suitable for machine learning algorithms? Explain why you made this choice.

Ans. If the categorical data has five unique values, I would use nominal encoding to transform the data into a format suitable 
for machine learning algorithms. Nominal encoding assigns a unique integer value to each category, which would allow the 
machine learning algorithm to process the data. In this case, one could assign the values 1, 2, 3, 4, and 5 to each of the 
five unique values.

Q5. In a machine learning project, you have a dataset with 1000 rows and 5 columns. Two of the columns are categorical, 
and the remaining three columns are numerical. If you were to use nominal encoding to transform the categorical data, how 
many new columns would be created? Show your calculations.

Ans. If we use nominal encoding to transform the two categorical columns into numerical data, we would create two new columns. 
Each categorical column would be replaced by a single numerical column. Therefore, the resulting dataset would have five numerical columns.

Q6. You are working with a dataset containing information about different types of animals, including their species, habitat,
and diet. Which encoding technique would you use to transform the categorical data into a format suitable for machine learning 
algorithms? Justify your answer.

Ans. In this case, I would use one-hot encoding to transform the categorical data into a format suitable for machine learning algorithms.
One-hot encoding would create a binary column for each category, which would indicate whether the animal belongs to a particular
category or not. For instance, one-hot encoding could create columns such as "species_cat," "species_dog," "habitat_forest," 
"habitat_ocean

Q7.You are working on a project that involves predicting customer churn for a telecommunicationscompany. You have a dataset 
with 5 features, including the customer's gender, age, contract type,monthly charges, and tenure. Which encoding 
technique(s) would you use to transform the categoricaldata into numerical data? Provide a step-by-step explanation of
how you would implement the encoding.

Ans. For this dataset, the categorical feature is the customer's contract type. We can transform this categorical data 
into numerical data using the following steps:

Create a new column for the encoded data: We can create a new column called "encoded_contract_type" to store the encoded data.

Assign a unique integer value to each category: We can assign a unique integer value to each category of the contract 
type. For example, we can assign "Month-to-month" a value of 1, "One year" a value of 2, and "Two year" a value of 3.

Encode the data: We can encode the data by replacing the categorical values in the "contract_type" column with the
corresponding integer values in the "encoded_contract_type" column.

After the above steps, the categorical data in the dataset would be transformed into numerical data, which would be suitable
for machine learning algorithms.
```

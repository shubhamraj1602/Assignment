```python
Q1. In order to predict house price based on several characteristics, such as location, square footage,number of bedrooms,
etc., you are developing an SVM regression model. Which regression metric in thissituation would be the best to employ?
Dataset link:https://drive.google.com/file/d/1Z9oLpmt6IDRNw7IeNcHYTGeJRYypRSC0/view?usp=share_link

Ans. In this situation, the best regression metric to employ for evaluating the performance of the SVM regression model would 
be the mean squared error (MSE). The reason for this is that the MSE measures the average of the squared differences between
the predicted and actual values, which provides a good measure of how well the model is able to accurately predict the house prices.
Additionally, the MSE is a commonly used metric in regression tasks and is relatively easy to interpret.

To compute the MSE, you would need to first split the dataset into training and test sets, fit the SVM regression model 
to the training set, predict the house prices for the test set using the trained model, and then compute the MSE between 
the predicted and actual house prices in the test set.

Q2. You have built an SVM regression model and are trying to decide between using MSE or R-squared asyour evaluation metric. 
Which metric would be more appropriate if your goal is to predict the actual price of a house as accurately as possible?

Ans. If the goal is to predict the actual price of a house as accurately as possible, then the evaluation metric that would 
be more appropriate is the mean squared error (MSE) rather than the R-squared metric.

The reason for this is that the MSE measures the average of the squared differences between the predicted and actual values, which
gives a good indication of how close the predicted values are to the actual values. On the other hand, the R-squared metric measures 
the proportion of the variance in the target variable that is explained by the model.

While R-squared can be a useful metric for assessing the overall fit of the model, it does not directly measure the accuracy of the
predicted values. In fact, a model with a high R-squared value may not necessarily be accurate in predicting individual target values. 
Therefore, for the specific goal of predicting the actual price of a house as accurately as possible, MSE would be the more appropriate 
evaluation metric to use.

Q3. You have a dataset with a significant number of outliers and are trying to select an appropriateregression metric 
to use with your SVM model. Which metric would be the most appropriate in this scenario?

Ans. When dealing with a dataset that contains a significant number of outliers, the most appropriate regression metric to use 
with an SVM model would be the mean absolute error (MAE).

The reason for this is that the MAE is less sensitive to outliers compared to other regression metrics such as mean squared error (MSE).
This is because the MAE measures the absolute differences between the predicted and actual values, which makes it less influenced by 
extreme values. On the other hand, the squared term in MSE can amplify the impact of outliers, making it less appropriate for datasets 
with a large number of outliers.

In summary, if your dataset contains a significant number of outliers, it would be best to use the MAE metric for evaluating the 
performance of your SVM regression model.

Q4. You have built an SVM regression model using a polynomial kernel and are trying to select the bestmetric to evaluate its
performance. You have calculated both MSE and RMSE and found that both valuesare very close. Which metric should you choose to use in this case?

Ans. If both the mean squared error (MSE) and root mean squared error (RMSE) values are very close to each other, either of the two metrics
could be used to evaluate the performance of the SVM regression model.

However, in practice, the RMSE is often preferred over MSE because it has the advantage of being in the same units as the target variable
(i.e., the dependent variable). This means that the RMSE is more interpretable compared to the MSE, as it provides a measure of the average 
deviation of the predicted values from the actual values in the same units as the target variable.

Therefore, if both MSE and RMSE values are very close, it would be appropriate to use the RMSE as the evaluation metric for the SVM regression model.

Q5. You are comparing the performance of different SVM regression models using different kernels (linear,polynomial, and RBF) 
and are trying to select the best evaluation metric. Which metric would be most appropriate if your goal is to measure how well 
the model explains the variance in the target variable?

Ans. If the goal is to measure how well the model explains the variance in the target variable, the most appropriate evaluation metric 
to use would be the coefficient of determination (R-squared).

The R-squared metric measures the proportion of the variance in the target variable that is explained by the model. Therefore, if the goal 
is to evaluate how well the SVM regression models using different kernels are able to explain the variance in the target variable, the R-squared
metric would be the most appropriate evaluation metric to use.

By comparing the R-squared values of the different models, you can determine which kernel performs best in explaining the variance in the target
variable. It is worth noting, however, that the R-squared metric has limitations, such as not being able to determine causation, so it is important 
to use it in conjunction with other evaluation metrics such as MSE and MAE to get a more comprehensive evaluation of the model's performance.
```
